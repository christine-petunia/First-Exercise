```python
# 导入库.
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

### 最简单的绘图例子
- matplotlib的图像绘制在figure上,figure中又包含坐标轴axes.`pyplot.subplots`命令可以快速创建figure和axes.

```python
fig, ax = plt.subplots() #创建子图和坐标轴
ax.plot([1,4,6,8],[2,5,3,4])
```

- `<matplotlib.lines.Line2D at 0x7fa3a82a1b90>`的出现是因为matplotlib的绘图代码默认打印出最后一个对象.解决方法有:
    - 在代码块最后加一个分号
    - 在代码块最后加一句`plt.show()`
    - 在绘图时将绘图对象显式赋值给一个变量,如`a = ax.plot([1,4,6,8],[2,5,3,4])`
- 上述代码也可以简化为`line = plt.plot([1,4,6,8],[2,5,3,4])`

### 分析figure的组成
- 对各类元素操作可以理解为在不同图层上操作,最后的图就是将这些图层整合的结果.
    - Figure:顶层级，用来容纳所有绘图元素
    - Axes:matplotlib宇宙的核心，容纳了大量元素用来构造子图
    - Axis:axes的下属层级，用于处理所有和坐标轴，网格有关的元素
    - Tick:axis的下属层级，用来处理所有和刻度有关的元素

- matplotlib提供两种最常用的绘图接口

```python
x = np.linspace(0, 2, 100) # 在线性空间中以均匀步长生成数字序列.
# 第一种:显式创建figure和axes，在上面调用绘图方法
fig, ax = plt.subplots()  
ax.plot(x, np.exp(x), label='quadratic')  
ax.plot(x, x**3+1, label='cubic')  
ax.set_xlabel('x label') 
ax.set_ylabel('y label') 
ax.set_title("Simple Plot")  
ax.legend() 
plt.show()
```

```python
# 第二种:依赖pyplot自动创建figure和axes并绘图
plt.plot(x, np.exp(x), label='quadratic') 
plt.plot(x, x**3+1, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
```

### 思考题
- 两种绘图模式的优缺点和各自适合的使用场景.
    - 面向对象方法:显示创建图像层和坐标层的好处是让我们对绘图过程有了完全的控制权,而且绘图的逻辑更强.
    - pyplot绘图模式隐式地创建了面向对象模式的相关对象.
- pyplot绘图模式的简单模板:

```python
# step1 准备数据
x = np.linspace(0, 2, 100)
y = x**2

# step2 设置绘图样式，这一模块的扩展参考第五章进一步学习，这一步不是必须的，样式也可以在绘制图像是进行设置
mpl.rc('lines', linewidth=4, linestyle='-.')

# step3 绘制图像， 这一模块的扩展参考第二章进一步学习
plt.plot(x, y, label='quadratic')

# step4 添加标签，文字和图例，这一模块的扩展参考第四章进一步学习
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.legend()
plt.show()
```
